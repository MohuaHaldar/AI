```
from PIL import Image
import pytesseract as ptc
from pdf2image import convert_from_path
import os
```

Write a function that will convert the pdf pages to image files

```
 
def convert_to_img(file_path):
    pages=convert_from_path(file_path,500)
    page_fol='/content/pages'
    if not os.path.exists(page_fol):
        os.mkdir(page_fol)
    img_ctr=1
 
    for page in pages:
       img_filenm='page_'+str(img_ctr)+'.jpg'
       page.save(os.path.join(page_fol,img_filenm),'JPEG')
       img_ctr+=1
```

```
 convert_to_img(file_path)
```

```
def convert_img_to_text(page_fol):
    page_list= os.listdir(page_fol)
    out_file=os.path.join(os.getcwd(),'outtext.txt')
    file=open(out_file,'a')
    for page in page_list:
      file_name=os.path.join(page_fol,page)
      txt=str(ptc.image_to_string(Image.open(file_name)))
      file.write(txt)
```

```
 convert_img_to_text(page_fol)
```
