```python
def remove_correlated_columns(df, thres):
    y=df['class']
    df=df.drop(columns=['class'], axis=1)
    corr_wine=df.corr()
    col_set=set()
    for i in range(len(corr_wine.columns)):
        for j in range(i):
            if (corr_wine.iloc[i,j]>thres and corr_wine.columns[j] not in col_set):
                col_name=corr_wine.columns[i]
                col_set.add(col_name)
                df.drop(columns=col_name, axis=1, inplace=True)
    df['class']=y
    return df
    
```

```python
wine=remove_correlated_columns(wine, .6)
```

```python
wine.shape
```

```python
# split into train and test
from sklearn.model_selection import train_test_split
target=wine['class']
features=wine.drop(columns=['class'])
X_train, X_test, y_train, y_test=train_test_split(features, target, test_size=.2, random_state=42)
```

```python
# scale the features
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
# Encode the target
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.fit_transform(y_test)
```

```python
from sklearn.model_selection import cross_val_score
def cross_val(X_train, y_train, model):
    acc=cross_val_score(estimator=model,X=X_train, y=y_train, cv=5)
    return acc.mean()
```

```python
def evaluate_model(model):
    model.fit(X_train, y_train)
    model_pred=model.predict(X_test)
    model_score=cross_val(X_train, y_train, model)
    return model_score
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr_score=evaluate_model(lr)
print('Logistic Regression score: ', lr_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn_score=evaluate_model(knn)
print('KNN Regression score: ', knn_score)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb_score=evaluate_model(nb)
print('NB Regression score: ', nb_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10, criterion='entropy')
rf_score=evaluate_model(rf)
print('Random forest score:',rf_score )
```

```python
from xgboost import XGBClassifier
gb=XGBClassifier()
gb_score=evaluate_model(gb)
print('XGBoost score: ', gb_score)
```

```python
plt.style.use('fivethirtyeight')
figsize=(8,6)
model_comp=pd.DataFrame({'model': ['LogisticRegression','KNN','Naive Bayes','Random ','Gradient Boosting'], 'score':[lr_score,knn_score,nb_score,rf_score,gb_score]})
model_comp.sort_values('score', ascending=True).plot(x='model',y='score', kind='barh', color='red', edgecolor='black')
plt.ylabel('')
plt.yticks(size=14)
plt.xlabel('K-fold cross validation')
plt.xticks(size=14)
plt.title('Model comparison on score', size=20)
```
