```
import pandas as pd
import numpy as np
```

```
sub=pd.read_csv('/content/sample_submission.csv')
sub.head()
```

```
tr_data=pd.read_csv(tr_path)
```

```
tr_data.head()
```

```
tr_text=tr_data['comment_text']
```

```
tst_data=pd.read_csv(tst_path)
tst_data.head()
```

```
tst_text=tst_data['comment_text']
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```
joint_data=pd.concat([tr_text,tst_text])
```

```
word_vec = TfidfVectorizer( sublinear_tf=True, strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}', ngram_range=(1, 1), max_features=30000)
```

```
word_vec.fit(joint_data)
```

```
 word_vec.get_feature_names()
```

```
word_vec.vocabulary_
```

```
tr_fea=word_vec.transform(tr_text)
```

```
tst_fea=word_vec.transform(tst_text)
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
rf=RandomForestClassifier()
```

```
labels=['toxic', 'severe_toxic', 'obscene', 'threat',
       'insult', 'identity_hate']
```

```
for label in labels:
     rf.fit(tr_fea,tr_data[label])
     tr_pred=rf.predict(tr_fea)
     print('Accuracy:', accuracy_score(tr_data[label],tr_pred))
     sub[label]=rf.predict_proba(tst_fea)[:,1]
```
