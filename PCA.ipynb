```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
import seaborn as sns
from sklearn import preprocessing
%matplotlib inline
```

```python
# Load the dataset
cancer_data=datasets.load_breast_cancer()
```

```python
# Load X and y
X=cancer_data.data
y=cancer_data.target
```

```python
X=preprocessing.scale(X)

X.shape
```

```python
# Perform the PCA
pca=PCA(30)
pca_full=pca.fit(X)
plt.plot(np.cumsum(pca_full.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pca=PCA(n_components=16)
pca_new=pca.fit(X)
```

```python
print(pca_new.explained_variance_)
print(pca_new.explained_variance_ratio_)
print(pca_new.explained_variance_ratio_.cumsum())
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pca=PCA(n_components=2)
pca=pca.fit_transform(X)
pca
```

```python
plt.figure(figsize=(15,10))
plt.scatter(pca[:,0], pca[:,1], c=y, edgecolor='none', alpha=.7, cmap=plt.get_cmap('jet', 10), s=20)
plt.colorbar()
```

```python
# Visualizing using 3 components
pca=PCA(n_components=3)
pca=pca.fit_transform(X)
```

```python
ax = plt.figure(figsize=(16,12)).gca(projection='3d')
ax.scatter(
    xs=pca[:,0], 
    ys=pca[:,1], 
    zs=pca[:,2], 
    c=y, 
    cmap='tab10'
)
ax.set_xlabel('pca one')
ax.set_ylabel('pca two')
ax.set_zlabel('pca three')
plt.show()
```

```python
# 2D plot between PCA1 and PCA3
plt.figure(figsize=(15,10))
plt.scatter(pca[:,0], pca[:,2], c=y, edgecolor='none', alpha=.7, cmap=plt.get_cmap('jet', 10), s=20)
plt.colorbar()
```

```python
# 2D plot between PCA2 and PCA3
plt.figure(figsize=(15,10))
plt.scatter(pca[:,1], pca[:,2], c=y, edgecolor='none', alpha=.7, cmap=plt.get_cmap('jet', 10), s=20)
plt.colorbar()
```
