```
import pandas as pd
```

```
data=pd.read_csv('/content/drive/My Drive/Amphi/dataset.csv',sep=';')
```

```
data.head()
```

```
import re
```

```
 
num_cols=[]
cat_cols=[]
for col in data.columns:
    if re.search(r'int|num',col,re.IGNORECASE):
       num_cols.append(data[col][0])
    if re.search(r'cat|lab', col, re.IGNORECASE):
       cat_cols.append(data[col][0])
```

```
print(num_cols)
print(cat_cols)
```

```
columns=list(data.iloc[0])
columns
```

```
data=data[1:]
```

```
data.head()
```

```
data.columns=columns
 
data.head()
```

```
for col in data.columns:
    print(col,':',data[col].nunique())
```

```
data.isnull().sum()
```

```
data.info()
```

```
for col in cat_cols:
    data[col]=data[col].astype('category')
for col in num_cols:
    data[col]=data[col].astype('int64')
```

```
data.info()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data.columns
labels=['Green frogs', 'Brown frogs',
       'Common toad', 'Fire-bellied toad', 'Tree frog', 'Common newt',
       'Great crested newt']
```

```
features=data.drop(columns=labels,axis=1)
features=features.drop(['ID','Motorway'], axis=1)
```

```
features.head()
```

```
lbl_data=data[labels]
lbl_data.head()
```

Rows having multiple labels

```
counts=[]
for col in lbl_data.columns:
    counts.append(lbl_data[col].value_counts()[0])
 
plt.figure(figsize=(12,10))
sns.barplot(lbl_data.columns, counts)
plt.show()
```

Great crested newt has the highest records

```
lbl_data.head()
```

```
 from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train, y_test=train_test_split(features, lbl_data, test_size=.2)
```

```
lr_pipe=Pipeline([( 'clf',OneVsRestClassifier(LogisticRegression(solver='liblinear'), n_jobs=1))])
```

```
 
 
for col in labels:
   lr_pipe.fit(X_train, y_train[col])
   y_pred=lr_pipe.predict(X_test)
   print('Accuracy:',accuracy_score(y_pred,y_test[col]))
```
